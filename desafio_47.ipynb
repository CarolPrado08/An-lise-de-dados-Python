Contando quantos alunos estavam presentes na aula do dia 24/08 na Turma H

```python
nomes_sem_repetir = tabela.drop_duplicates(subset='Nome')


alunos_presentes = len(nomes_sem_repetir)

print(f'Quantidade de alunos presentes no dia 24/08: {alunos_presentes}')

novo_dataframe = nomes_sem_repetir[['Nome', 'Hora', 'Atividade']]


novo_dataframe.to_excel('arquivo_alunos_presentes.xlsx', index=False)

planilha_presentes = pd.read_excel("arquivo_alunos_presentes.xlsx")
display(planilha_presentes)
```
